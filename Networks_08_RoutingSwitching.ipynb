## Introduction to Routing and Switching Technologies

Modern computer networks rely on two fundamental technologies to move data from source to destination: routing and switching. These complementary technologies work at different layers of the OSI model but collaborate to create efficient and reliable network communications. **Routing** refers to the process of determining the best path for data to travel across multiple networks, while **switching** focuses on forwarding data packets between devices within the same network.

### Key Concepts in Routing and Switching:
* **Routers** operate at Layer 3 (Network Layer) of the OSI model and make decisions based on logical addressing (IP addresses) to forward packets between different networks.
* **Switches** function at Layer 2 (Data Link Layer) and use physical addressing (MAC addresses) to forward frames within the same network segment.
* Routing and switching technologies work together to create scalable, efficient, and reliable network infrastructures that support modern applications and services.
* Understanding both technologies is essential for designing, implementing, and troubleshooting computer networks of all sizes.

### Components of Network Data Flow:
1. When data is sent across a network, it first moves through the local network via switches.
2. If the destination is on a different network, the data is forwarded to a router that determines the best path to reach that network.
3. The data may pass through multiple routers and switches before reaching its final destination.
4. Each routing and switching decision is made independently based on current network conditions and configured protocols.

In this chapter, we'll explore the various routing protocols, switching technologies, and configuration options that network engineers use to create efficient and resilient networks. By understanding these concepts, you'll gain insight into how data travels across both local and global networks.

## Understanding Static vs. Dynamic Routing

Network routing can be implemented through two primary methods: static and dynamic routing. Each approach has distinct characteristics that make it suitable for different network environments and requirements. **Static routing** involves manually configuring fixed routes between networks, while **dynamic routing** uses protocols that automatically discover and adjust to network changes.

### Static Routing Characteristics:

Static routing offers predictability and simplicity but requires manual configuration. Key aspects of static routing include:

* **Static routes** are manually configured by network administrators and remain unchanged until explicitly modified.
* They create predictable and consistent network paths that are not influenced by network conditions or traffic patterns.
* Static routing consumes minimal router resources since no routing protocol calculations are needed.

This approach is particularly beneficial in certain scenarios:

* Small, stable networks with few alternative paths
* Backup routes in case dynamic routing fails
* Lab environments where predictable routing behavior is desired
* The administrative overhead increases significantly as the network grows, making static routing impractical for large or complex networks.

### Dynamic Routing Characteristics:

Dynamic routing provides automation and adaptability that static routing cannot match. The primary features of dynamic routing include:

* **Dynamic routing protocols** automatically discover network topology and exchange routing information between routers.
* They continuously adapt to network changes, failures, and congestion by recalculating optimal paths.
* Dynamic routing scales effectively for large networks as it requires minimal manual configuration when adding new networks.

While offering significant advantages, dynamic routing also has considerations to keep in mind:

* These protocols consume more router resources (CPU, memory, bandwidth) than static routing due to constant calculations and updates.
* Dynamic routing provides fault tolerance and load balancing capabilities that static routing cannot offer.
* Initial configuration is often more complex than static routing, requiring understanding of protocol-specific parameters.

| Characteristic | Static Routing | Dynamic Routing |
|----------------|---------------|-----------------|
| Configuration | Manual | Automatic |
| Scalability | Poor | Excellent |
| Resource Usage | Minimal | Moderate to High |
| Adaptation to Changes | None (manual updates required) | Automatic |
| Administrative Overhead | High in large networks | Low once configured |
| Best Use Case | Small networks, backup routes | Medium to large networks |

The choice between static and dynamic routing often depends on network size, complexity, and requirements for adaptability. Many modern networks use a combination of both approaches, with static routes for certain stable connections and dynamic routing protocols for the broader network infrastructure.

## Dynamic Routing Protocols: BGP, EIGRP, and OSPF

Dynamic routing protocols enable routers to automatically discover network destinations and adapt to topology changes without manual intervention. These protocols differ in their operation, scalability, and intended use cases. Let's explore three important dynamic routing protocols: BGP, EIGRP, and OSPF.

### Border Gateway Protocol (BGP):

**BGP** serves as the routing foundation of the global internet, with specialized characteristics that set it apart from interior routing protocols:

* It is the primary routing protocol of the internet, designed to exchange routing information between different autonomous systems (AS).
* BGP is classified as a path-vector protocol that makes routing decisions based on network policies, rule sets, and path attributes rather than just metrics.
* It focuses on stability and policy enforcement rather than finding the fastest path, making it suitable for internet service providers and large enterprise networks.

BGP's specialized design addresses the unique challenges of internet-scale routing:
* It is highly scalable but more complex to configure than interior gateway protocols, requiring careful planning and skilled administration.
* BGP uses TCP port 179 for reliable communication between BGP peers and maintains connections through keepalive messages.
* It can filter and manipulate routes based on complex policies that reflect business relationships and traffic engineering requirements.

### Enhanced Interior Gateway Routing Protocol (EIGRP):

**EIGRP** provides efficient routing within enterprise networks through its advanced distance-vector approach:

* It is a Cisco proprietary advanced distance-vector routing protocol that combines the best features of both distance-vector and link-state protocols.
* EIGRP uses the Diffusing Update Algorithm (DUAL) to calculate the shortest path to destinations and maintain backup routes for quick convergence.
* It is efficient with network resources, sending updates only when changes occur and only to the routers that need the information.

EIGRP offers several advantages that make it popular in many enterprise environments:
* It supports multiple network layer protocols and unequal cost load balancing, making it versatile for various network designs.
* While originally proprietary to Cisco, limited features were later released as an open standard, but full functionality remains on Cisco devices.
* EIGRP provides faster convergence times than traditional distance-vector protocols while consuming fewer resources than link-state protocols.

### Open Shortest Path First (OSPF):

**OSPF** is a standards-based link-state protocol that creates a complete map of the network topology:

* It is an open standard link-state routing protocol widely used in large enterprise networks.
* OSPF divides the network into areas to improve scalability and uses the Dijkstra algorithm to calculate the shortest path to destinations.
* It creates a complete topology map of the network and updates it when changes occur, ensuring all routers have identical views of the network.

OSPF provides a structured approach to network routing:
* It converges quickly after network changes and supports various network types, including broadcast, point-to-point, and non-broadcast multi-access.
* OSPF uses a hierarchical design with a backbone area (Area 0) connecting to other areas, which reduces routing update traffic and processing requirements.
* As an open standard, it works across devices from different vendors, making it ideal for multi-vendor environments.

| Protocol | Type | Best Use Case | Convergence Speed | Scalability | Complexity |
|----------|------|--------------|------------------|-------------|------------|
| BGP | Path-Vector | Internet, Service Providers | Slow | Very High | High |
| EIGRP | Advanced Distance-Vector | Enterprise Networks | Very Fast | Medium | Medium |
| OSPF | Link-State | Enterprise Networks | Fast | High | Medium-High |

When selecting a dynamic routing protocol, network administrators must consider factors such as network size, topology, convergence requirements, and vendor support. Many large networks implement multiple routing protocols, with each serving specific parts of the network based on their strengths.

## Route Selection Criteria: Administrative Distance, Prefix Length, and Metrics

When multiple routing protocols or routes to the same destination exist, routers must determine which path to use. Routers employ a systematic approach to route selection based on three primary criteria: administrative distance, prefix length, and routing metrics. Understanding these criteria is essential for predicting and controlling network traffic flow.

### Administrative Distance:

The first factor in route selection is administrative distance, which determines which routing protocol to trust when multiple protocols advertise the same network:

* **Administrative distance (AD)** is a value that indicates the trustworthiness of a routing information source, with lower values being more preferred.
* When a router learns about a destination network from multiple routing protocols, it selects the route with the lowest administrative distance.

Administrative distance functions as a router's way of prioritizing routing information sources:

* Administrative distance values are locally significant on the router and are not exchanged between routers.
* This value helps resolve conflicts between different routing protocols and ensures consistent routing decisions.
* Administrators can modify default AD values to influence routing decisions based on specific network requirements.

### Prefix Length:

After evaluating administrative distance, the router examines the specificity of routes using prefix length:

* **Prefix length** refers to the number of bits in the subnet mask that are set to 1, often written in CIDR notation (e.g., /24).
* When multiple routes to the same destination exist with equal administrative distances, the router selects the route with the longest prefix match (most specific route).

This concept follows the principle that more specific information should take precedence:

* Longer prefix lengths indicate more specific routes (e.g., /24 is more specific than /16), providing more precise targeting of network destinations.
* This approach, known as **longest prefix match**, ensures that traffic is routed based on the most specific information available.
* It allows for hierarchical routing, where general routes cover large areas and specific routes provide targeted path selection.

### Routing Metrics:

The final criterion applied is the routing metric, which varies by protocol but always measures path desirability:

* **Metrics** are values used by routing protocols to determine the best path when multiple routes with the same administrative distance and prefix length exist.

Different protocols calculate metrics using different parameters:

* OSPF uses **cost**, which is based on bandwidth by default (lower is better)
* EIGRP uses a composite metric based on bandwidth, delay, reliability, load, and MTU (lower is better)
* RIP uses **hop count**, which is the number of routers that must be traversed (lower is better)

These metrics allow routing protocols to select optimal paths based on network conditions and administrator preferences. By adjusting metric calculations, network administrators can influence traffic patterns to maximize performance or meet specific business requirements.

| Route Selection Criterion | Priority | Example Values | Used When |
|---------------------------|----------|----------------|-----------|
| Administrative Distance | 1st | Connected: 0<br>Static: 1<br>EIGRP: 90<br>OSPF: 110<br>RIP: 120 | Multiple routing protocols advertise the same destination |
| Prefix Length | 2nd | /32 (most specific)<br>/24<br>/16<br>/8 (least specific) | Multiple routes with the same AD exist |
| Routing Metric | 3rd | Varies by protocol<br>(lower is better) | Multiple routes with the same AD and prefix length exist |

Route selection process is hierarchical and deterministic. The router first compares administrative distances, then prefix lengths, and finally metrics. Only when routes tie at one level does the router proceed to the next criterion. This systematic approach ensures that routers make consistent and predictable forwarding decisions across the network.

## Network Address Translation (NAT) and Port Address Translation (PAT)

Address translation technologies solve the problem of IPv4 address scarcity and provide additional security benefits by hiding internal network addresses. These technologies allow multiple devices to share a single or small pool of public IP addresses. **Network Address Translation (NAT)** and **Port Address Translation (PAT)** are the two primary methods used to implement address translation in modern networks.

### Network Address Translation (NAT):

**NAT** fundamentally alters how devices with private IP addresses communicate with public networks:

* It is the process of modifying IP address information in packet headers while in transit across a routing device.
* NAT enables private IP addresses (RFC 1918 addresses) to communicate with public networks by translating them to public IP addresses.

There are several common implementations of NAT, each serving different network requirements:

* **Static NAT**: Maps a single private IP address to a single public IP address in a fixed one-to-one relationship
* **Dynamic NAT**: Maps a private IP address to a public IP address from a pool of available addresses
* **NAT Overload (PAT)**: Maps multiple private IP addresses to a single public IP address using different port numbers

### Port Address Translation (PAT):

**PAT** extends basic NAT functionality by incorporating port numbers into the translation process:

* Also known as NAT Overload, it extends basic NAT by using port numbers to distinguish between different connections.
* With PAT, multiple internal devices can share a single public IP address, with the router keeping track of connections using source port numbers.

The PAT process involves a sophisticated tracking mechanism to maintain connection state:

* When an internal device initiates a connection, the router:
  1. Records the original source IP address and port in a translation table
  2. Replaces them with the public IP address and a unique port number
  3. Forwards the packet to its destination
  4. Reverses the process for returning traffic

This approach has become ubiquitous in small networks:
* PAT is the most common form of NAT in home and small business networks due to its efficiency in conserving public IP addresses.
* It allows hundreds or even thousands of internal devices to share a single public IP address.

### Benefits and Limitations of Address Translation:

Address translation technologies offer multiple advantages, but also come with certain trade-offs that must be considered.

The key benefits include:
* Conservation of public IPv4 addresses by allowing many devices to share one or few public IPs
* Enhanced security by hiding internal network addressing from external networks
* Simplified internal addressing schemes that don't need to change when external connections change
* Basic firewall functionality as unsolicited incoming connections are typically blocked

However, these technologies also introduce challenges:
* Some applications and protocols don't work well with NAT and may require special configuration
* End-to-end connectivity and traceability are compromised, making troubleshooting more complex
* Additional processing overhead on routing devices can affect performance in high-traffic environments
* IPv6 reduces the need for NAT due to its vastly larger address space

|---------|-----------|-------------|-------------------|
| Public IP Addresses Required | One per internal host | Equal to maximum simultaneous users | As few as one for many users |
| Translation Type | 1:1 (one private to one public) | Many:Many (from address pool) | Many:1 (many private to one public) |
| Port Translation | No | No | Yes |
| Internal Host Accessibility from Outside | Yes, consistently | Yes, when translation is active | Requires additional configuration |
| Common Use Case | Public servers behind NAT | Medium networks with occasional external access | Home and business internet connections |

Address translation remains a critical technology in today's networks despite the gradual transition to IPv6. Understanding how NAT and PAT work is essential for network design, troubleshooting, and security implementation.

## First Hop Redundancy Protocol (FHRP) and Virtual IP Addresses

Network availability is critical for modern businesses. First Hop Redundancy Protocols (FHRPs) provide a way to ensure that if a default gateway (router) fails, another device can take over automatically. This redundancy happens so quickly that users don't even notice the change.

### First Hop Redundancy Protocol Concepts:

**First Hop Redundancy Protocols (FHRPs)** create the illusion of a single, reliable gateway by coordinating between multiple physical routers:

* FHRPs allow multiple routers to work together to provide a fault-tolerant default gateway
* They share a single virtual IP address that serves as the default gateway for computers on a network
* If the primary router fails, a backup router automatically takes over

This automatic failover is essential for businesses that need constant network connectivity.

### Virtual IP (VIP) Addresses:

A **Virtual IP (VIP)** address is a logical address shared among multiple devices:

* It's not tied to a specific physical device but can move between devices as needed
* End devices (like computers) are configured to use this virtual IP as their default gateway
* This approach means end devices don't need to be reconfigured when a physical router fails

Virtual IP addresses provide the foundation for router redundancy by creating a consistent point of contact that survives hardware failures.

### Common FHRP Implementations:

There are three main types of First Hop Redundancy Protocols:

1. **Hot Standby Router Protocol (HSRP)**:
   * Cisco's protocol that uses an active router and one or more standby routers
   * Only the active router forwards traffic, while standby routers wait to take over if needed

2. **Virtual Router Redundancy Protocol (VRRP)**:
   * An industry standard that works across different vendors' equipment
   * Functions similarly to HSRP with master and backup routers

3. **Gateway Load Balancing Protocol (GLBP)**:
   * Provides both redundancy and load balancing
   * Allows multiple routers to actively forward traffic at the same time
   * Distributes the network load across all available routers

| Feature | HSRP | VRRP | GLBP |
|---------|------|------|------|
| Vendor Support | Cisco only | Multi-vendor | Cisco only |
| Active Routers | 1 active | 1 master | Multiple active |
| Load Balancing | No | No | Yes |

Implementing FHRPs is essential in business networks where continuous connectivity is required. These protocols ensure that if a router fails, network traffic continues to flow through an alternate path with minimal disruption.

## Subinterfaces: Configuration and Use Cases

Router interfaces can be logically divided into multiple virtual interfaces, each operating as if it were a separate physical interface. These logical divisions, called **subinterfaces**, allow a single physical connection to serve multiple network functions. Subinterfaces are particularly valuable in environments where physical router ports are limited but multiple logical networks need to be supported.

### Understanding Subinterfaces:
* A **subinterface** is a logical division of a physical interface that operates independently with its own network settings.
* Subinterfaces are identified by the physical interface name followed by a period and a number (e.g., GigabitEthernet0/0.10).
* Each subinterface requires its own IP address and subnet mask, functioning as a distinct router interface.
* Subinterfaces are commonly used with VLAN trunking to allow a single physical connection to route traffic for multiple VLANs.

### Primary Use Cases for Subinterfaces:

Router-on-a-Stick configuration is one of the most common applications for subinterfaces. This approach allows a single router interface to route traffic between multiple VLANs.

The key elements of this configuration include:
* A physical connection between the router and a switch configured as a trunk port
* Multiple subinterfaces on the router, each mapped to a different VLAN
* 802.1Q encapsulation on each subinterface to identify VLAN traffic

Other important use cases include:
* **WAN connections**: Subinterfaces can be used on WAN links to support multiple virtual circuits or connections over a single physical interface.
* **Service provider connections**: When connecting to a service provider that uses VLAN tagging to separate customer traffic, subinterfaces allow proper traffic segregation.
* **Network segmentation**: Creating separate broadcast domains without requiring additional physical interfaces, saving hardware costs and rack space.

### Basic Subinterface Configuration:

The configuration of a subinterface typically includes these essential parameters:
* The interface identifier with a subinterface number
* Encapsulation type (most commonly dot1q for 802.1Q VLAN tagging)
* VLAN identifier
* IP address and subnet mask

```
Router(config)# interface GigabitEthernet0/0.10
Router(config-subif)# encapsulation dot1q 10
Router(config-subif)# ip address 192.168.10.1 255.255.255.0
```

| Configuration Element | Description | Example |
|----------------------|-------------|---------|
| Physical Interface | The actual hardware port on the router | GigabitEthernet0/0 |
| Subinterface Designation | Number that identifies the specific subinterface | GigabitEthernet0/0.10 |
| Encapsulation | Protocol used to identify different traffic types | encapsulation dot1q 10 |
| Native VLAN | Untagged VLAN on a trunk (optional) | encapsulation dot1q 10 native |
| IP Address | Network address for the subinterface | ip address 192.168.10.1 255.255.255.0 |

Subinterfaces provide flexibility in network design with minimal hardware requirements. They allow network administrators to create complex routing configurations while conserving physical ports. However, it's important to note that all subinterfaces share the bandwidth of the physical interface, which can become a bottleneck in high-traffic environments.

## Virtual Local Area Networks (VLANs) and Switch Virtual Interfaces (SVIs)

In a traditional network, all devices connected to the same switch are part of the same broadcast domain and can communicate directly with each other. However, this arrangement can create performance and security challenges as networks grow. **Virtual Local Area Networks (VLANs)** solve these problems by dividing a physical network into logical segments.

### Understanding VLANs:

Think of a VLAN as a way to create separate "virtual switches" within a single physical switch:

* A **VLAN** is a logical grouping of network devices that behave as if they're connected to the same physical network, even if they're connected to different ports or different switches
* Each VLAN creates a separate broadcast domain, meaning broadcast messages only reach devices within the same VLAN
* Devices in different VLANs cannot communicate directly with each other without the help of a router

To understand this concept, imagine a school with different departments. Using VLANs is like creating separate networks for the administration office, the science department, and the libraryeven though they all share the same physical network infrastructure.

### Benefits of VLANs:

VLANs provide several important advantages in network design:

* **Improved security**: Traffic is isolated between VLANs, limiting potential security breaches
* **Better performance**: Broadcast traffic is contained within each VLAN, reducing network congestion
* **Simplified management**: Devices can be grouped logically by function, department, or application rather than by physical location
* **Cost savings**: A single physical network can support multiple logical networks, reducing hardware costs

### VLAN Basics:

VLANs are identified by numbers. On most switches:

* VLAN 1 is the default VLAN where all ports initially belong
* VLANs 1-1001 are the standard VLAN range
* Each VLAN is typically assigned its own subnet (range of IP addresses)

When setting up VLANs, you might create something like:
* VLAN 10: Administration (192.168.10.0/24)
* VLAN 20: Faculty (192.168.20.0/24)
* VLAN 30: Students (192.168.30.0/24)

### Switch Virtual Interfaces (SVIs):

When devices in different VLANs need to communicate, they require routing. In many modern networks, this routing happens right on the switch through **Switch Virtual Interfaces (SVIs)**:

* An **SVI** is a virtual interface on a switch that provides Layer 3 (routing) functionality for a VLAN
* Each SVI corresponds to a specific VLAN and acts as a gateway for devices in that VLAN
* SVIs allow communication between different VLANs without requiring a separate physical router

Using our school example, an SVI for each department VLAN would allow a teacher to access resources in the administration VLAN while maintaining the separation between networks.

### Real-World VLAN Applications:

VLANs are used in nearly all business networks for various purposes:

* **Separating voice and data traffic**: Many organizations place IP phones in a separate VLAN from computers
* **Guest access**: Visitor Wi-Fi can be placed in a separate VLAN with limited access to internal resources
* **IOT devices**: Smart building systems (cameras, thermostats, etc.) are often isolated in their own VLAN for security
* **Different security zones**: Public-facing servers might be in a different VLAN than internal databases

Understanding VLANs and SVIs is fundamental to modern network design. These technologies allow network administrators to create flexible, secure, and efficient network architectures that can grow and change with the organization's needs.
* They enable network management access to the switch itself
* They allow Layer 3 switches to perform inter-VLAN routing without an external router

SVI configuration requires:
* Creating the VLAN first
* Assigning the VLAN to physical ports
* Creating and configuring the SVI with an IP address that will serve as the default gateway for the VLAN

Basic SVI configuration:
```
Switch(config)# vlan 10
Switch(config-vlan)# name MANAGEMENT
Switch(config-vlan)# exit
Switch(config)# interface vlan 10
Switch(config-if)# ip address 192.168.10.1 255.255.255.0
Switch(config-if)# no shutdown
```

| VLAN Type | Range | Description | Use Case |
|-----------|-------|-------------|----------|
| Normal | 1-1001 | Standard VLANs | General network segmentation |
| Extended | 1006-4094 | Extended range VLANs | Large enterprise environments |
| Reserved | 1002-1005 | Reserved for legacy protocols | Rarely used in modern networks |
| Default | 1 | Default VLAN | All ports belong to this VLAN by default |
| Management | Typically VLAN 1 or a dedicated VLAN | For switch management | Network administration access |
| Native | Configurable (default: VLAN 1) | Untagged VLAN on trunk links | For legacy devices without VLAN tagging support |

VLANs and SVIs are fundamental components of modern network design. ## Interface Configuration Essentials: VLANs, Tagging, and Link Aggregation

Once you understand the concept of VLANs, the next step is learning how switches handle VLAN traffic through different port types and how multiple links can work together. These concepts form the foundation of practical networking.

### Access Ports vs. Trunk Ports:

Switch ports that handle VLAN traffic can be configured in two main ways:

* **Access Ports**:
  * These are the simplest type of switch port
  * Each access port belongs to exactly one VLAN
  * When a device (like a computer or printer) connects to an access port, it becomes part of that VLAN
  * The device doesn't need to know anything about VLANs - the switch handles everything

* **Trunk Ports**:
  * These special ports can carry traffic for multiple VLANs
  * They're primarily used for connections between switches or between a switch and a router
  * Trunk ports use a method called "tagging" to keep track of which frame belongs to which VLAN

Think of access ports like local roads that only serve one neighborhood, while trunk ports are like highways that connect multiple neighborhoods together.

### VLAN Tagging and 802.1Q:

When traffic from multiple VLANs travels over a single trunk link, the switch needs a way to identify which VLAN each frame belongs to. This is done through **VLAN tagging**:

* **802.1Q** is the industry standard for VLAN tagging
* It works by inserting a small tag (4 bytes) into the Ethernet frame header
* This tag contains a VLAN ID that identifies which VLAN the frame belongs to
* When a frame arrives on a trunk port, the switch reads the tag to determine which VLAN it should be delivered to

One special VLAN on a trunk is called the **native VLAN**. Traffic on this VLAN isn't tagged, which helps with compatibility for older devices that don't understand VLAN tags.

### Voice VLANs:

A **voice VLAN** is a special configuration that allows a single switch port to support both a phone and a computer:

* The phone is placed in one VLAN (typically the voice VLAN)
* A computer connected to the phone's PC port is placed in a different VLAN (typically the data VLAN)
* This separation allows the network to prioritize voice traffic for better call quality
* The switch automatically detects the phone and assigns it to the voice VLAN

This configuration is very common in office environments and demonstrates how a single physical port can support multiple logical networks.

### Link Aggregation:

Sometimes a single network connection between switches isn't enough. **Link aggregation** addresses this limitation:

* **Link aggregation** (also called port channeling or EtherChannel) combines multiple physical connections into one logical connection
* For example, four 1 Gbps links can be combined to create a single 4 Gbps logical link
* This provides two key benefits:
  * **Increased bandwidth**: More data can flow between the switches
  * **Redundancy**: If one physical link fails, traffic continues to flow through the remaining links

Link aggregation is like adding more lanes to a highway instead of being limited to a single lane in each direction.

### Speed and Duplex:

Two fundamental settings for any network interface are **speed** and **duplex**:

* **Speed** refers to how fast data can be transmitted (typically 10/100/1000 Mbps for Ethernet)
* **Duplex** refers to the direction of data flow:
  * **Half-duplex**: Data can flow in both directions, but only one direction at a time (like a walkie-talkie)
  * **Full-duplex**: Data can flow in both directions simultaneously (like a telephone)

Most modern networks operate at the highest available speed and in full-duplex mode for maximum performance.

Understanding these foundational switching concepts helps network administrators design and implement efficient, scalable networks that meet the needs of modern organizations.

Configuration of trunk links involves specifying both the tagging protocol and native VLAN:

| Configuration | Use Case | Example Command |
|---------------|----------|----------------|
| Access Port | End-user connection | `switchport mode access` <br> `switchport access vlan 10` |
| Trunk Port | Switch-to-switch connection | `switchport mode trunk` <br> `switchport trunk allowed vlan 10,20,30` |
| Native VLAN | Untagged VLAN on trunk | `switchport trunk native vlan 99` |
| Voice VLAN | IP phone connection | `switchport voice vlan 20` |
| Link Aggregation | Bandwidth increase/redundancy | `channel-group 1 mode active` |
| Speed/Duplex | Performance optimization | `speed 1000` <br> `duplex full` |

Proper interface configuration ensures efficient data flow, appropriate segregation of traffic, and optimal use of network resources. Understanding these essential configuration elements is fundamental for network administrators to build secure and high-performing network infrastructures.

## Spanning Tree Protocol: Preventing Network Loops

In network design, redundancy is importanthaving backup connections helps ensure that if one path fails, another is available. However, this redundancy creates a potential problem: **network loops**. The **Spanning Tree Protocol (STP)** was developed to solve this critical issue.

### The Problem: Network Loops

To understand why network loops are dangerous, imagine what happens when a computer sends a broadcast message (a message meant for all devices) on a network with a loop:

1. The broadcast is sent out from the computer to the first switch
2. The switch forwards the broadcast out all ports (as all broadcasts are handled)
3. The next switch does the same thing, sending the broadcast back to the first switch
4. The first switch, not remembering it already forwarded this broadcast, sends it again
5. This cycle continues indefinitely, creating a "broadcast storm"

This situation can quickly overwhelm a network, causing it to slow dramatically or even crash completely. Other problems caused by loops include:

* Multiple copies of the same frame arriving at the destination
* MAC address table instability (switches become confused about which port leads to which device)
* Excessive CPU usage on network devices

### The Solution: Spanning Tree Protocol

**Spanning Tree Protocol (STP)** prevents loops while still allowing for redundant connections. It works like this:

* All switches in the network communicate with each other
* They collectively create a logical tree-like structure (hence the name "spanning tree")
* Some network paths remain active while others are temporarily blocked
* If an active path fails, a blocked path can be unblocked to restore connectivity

It's like having multiple roads between locations but putting some roads on "standby," only opening them when the main road is closed.

### How STP Works:

STP follows a structured process to create a loop-free topology:

1. **Root Bridge Election**: All switches participate in electing one switch to be the "root" of the spanning tree. This switch becomes the reference point for the entire network.

2. **Path Cost Calculation**: Each switch determines its shortest path to the root bridge.

3. **Port Selection**: Based on these calculations, each switch determines the status of its ports:
   * **Root ports**: The best path to the root bridge
   * **Designated ports**: The best path from the root bridge to each network segment
   * **Blocked ports**: Ports that would create loops and are therefore disabled for regular traffic

When a network change occurs (like a link failure), STP recalculates and can unblock previously blocked ports to restore connectivity.

### STP in Action: A Simple Example

Imagine a network with three switches connected in a triangle:

* Switch A is elected as the root bridge
* Switch B and C both have direct connections to Switch A, so they each establish their root port toward A
* Switches B and C are also connected to each other
* STP will block one end of the B-to-C connection to prevent a loop
* If the direct connection between Switch A and Switch B fails, STP will unblock the B-to-C connection, allowing B to reach A through C

### Enhancements to STP

The original STP standard (IEEE 802.1D) has been improved over time:

* **Rapid Spanning Tree Protocol (RSTP)**: Provides much faster convergence when the network topology changes
* **Multiple Spanning Tree Protocol (MSTP)**: Allows different spanning trees for different groups of VLANs
* **Per-VLAN Spanning Tree Plus (PVST+)**: Runs a separate spanning tree instance for each VLAN

Most modern networks use one of these enhanced versions of STP, but the basic principle remains the same: prevent loops while maintaining redundancy.

Understanding Spanning Tree Protocol is essential for anyone working with switched networks. It quietly operates in the background of nearly every enterprise network, preventing the chaos that would otherwise result from redundant connections.

| Feature | Original STP (802.1D) | RSTP (802.1w) | MSTP (802.1s) |
|---------|----------------------|--------------|---------------|
| Convergence Time | 30-50 seconds | 1-5 seconds | 1-5 seconds |
| Port States | 5 states | 3 states | 3 states |
| Instances | Single spanning tree | Single spanning tree | Multiple instances |
| VLAN Support | One tree for all VLANs | One tree for all VLANs | Multiple VLANs per instance |
| Resource Usage | Low | Medium | Varies |
| Complexity | Low | Medium | High |

Spanning Tree Protocol remains a fundamental technology in switched networks, ensuring a loop-free topology while maintaining network resilience. Understanding STP operation is crucial for network administrators to design reliable networks and troubleshoot connectivity issues that may arise from spanning tree operations.

## Maximum Transmission Unit (MTU) and Jumbo Frames

Network performance optimization often involves adjusting the size of data packets traversing the network. Two related concepts that play a significant role in this optimization are the **Maximum Transmission Unit (MTU)** and **Jumbo Frames**. Understanding these concepts helps network administrators balance efficiency, reliability, and compatibility in their network designs.

### Maximum Transmission Unit (MTU):

The **Maximum Transmission Unit (MTU)** defines the largest packet size that can be transmitted as a single unit across a network link. This fundamental networking parameter affects both performance and reliability:

* MTU is measured in bytes and represents the maximum frame size at the data link layer
* The standard Ethernet MTU is 1500 bytes, which has been the default since early Ethernet implementations
* When a packet exceeds the MTU of a link, it must be fragmented (divided into smaller packets) or dropped
* Different network technologies have different default MTU sizes:
  * Ethernet: 1500 bytes
  * PPP: 1500 bytes
  * Token Ring: 4464 bytes
  * FDDI: 4352 bytes

### Impact of MTU Size on Network Performance:

MTU size affects network performance in several important ways:

* **Larger MTU values**:
  * Reduce overhead by requiring fewer packets for the same amount of data
  * Decrease CPU utilization on routers and endpoints
  * Lower overall latency for large data transfers
  * May increase transmission delays for time-sensitive small packets

* **Smaller MTU values**:
  * Reduce the impact of packet loss (less data needs to be retransmitted)
  * Decrease transmission delays for individual packets
  * Can help avoid fragmentation in networks with varied path MTUs
  * Increase overhead due to more packet headers for the same amount of data

### Jumbo Frames:

**Jumbo frames** are Ethernet frames with an MTU larger than the standard 1500 bytes:

* Typically, jumbo frames have an MTU of 9000 bytes, though values between 1501 and 9216 bytes may be used
* They were developed to enhance performance in high-speed networks, particularly Gigabit Ethernet and faster
* Jumbo frames are especially beneficial for large data transfers such as file storage, backup operations, and video streaming
* They're commonly implemented in data centers, storage area networks (SANs), and server farms

### Implementing Jumbo Frames:

To successfully implement jumbo frames, several requirements must be met:

* All devices in the communication path must support the same jumbo frame size
* End devices, switches, and routers must be configured with matching MTU values
* The entire path between communicating devices must support the larger frame size
* Basic configuration example:

```
Switch(config)# system mtu jumbo 9000
Switch(config)# exit
Switch# reload
```

| Consideration | Standard MTU (1500 bytes) | Jumbo Frames (9000 bytes) |
|---------------|--------------------------|---------------------------|
| Compatibility | Universal support | Limited to compatible devices |
| CPU Overhead | Higher (more packets) | Lower (fewer packets) |
| Network Efficiency | Lower for large transfers | Higher for large transfers |
| Fragmentation Risk | Higher across varied networks | Higher if path doesn't fully support |
| Best Use Case | General-purpose networks, Internet | Storage networks, data centers |
| Impact of Packet Loss | Lower (less data per packet) | Higher (more data per packet) |

### Path MTU Discovery:

**Path MTU Discovery (PMTUD)** is a technique that helps identify the optimal MTU size for a communication path:

* It works by sending packets with the "Don't Fragment" (DF) bit set
* If a router cannot forward the packet due to MTU constraints, it returns an ICMP "Fragmentation Needed" message
* The source then reduces its MTU and attempts transmission again
* This process continues until a suitable MTU is found for the entire path

Understanding MTU settings and properly implementing jumbo frames when appropriate can significantly enhance network performance, especially for applications involving large data transfers. However, compatibility issues must be carefully considered to avoid fragmentation problems and ensure reliable network operation.

````python
# @title
%%html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>Layer 3 Switch Decision Challenge</title>
  <style>
    body { font-family: sans-serif; margin: 20px; }
    h1, h2 { margin-bottom: 10px; }
    form { margin-top: 20px; }
    label { display: block; margin-top: 10px; }
    input, select, button { padding: 5px; font-size: 1em; margin-top: 5px; }
    table { border-collapse: collapse; margin-top: 20px; }
    table, th, td { border: 1px solid #ccc; padding: 8px; }
  </style>
</head>
<body>
  <h1>Layer 3 Switch Decision Challenge</h1>
  <p>
    In this exercise a <strong>Packet</strong> is provided along with a set of rules. Consult the <strong>VLAN Database</strong> to determine which VLAN a given <strong>MAC Address</strong> belongs to and decide whether the packet should be <strong>forwarded</strong> (if the destination is on the same VLAN), <strong>routed</strong> (if it belongs to a different VLAN and inter-VLAN routing applies), <strong>dropped</strong> (if the ingress VLAN is invalid), or <strong>broadcast</strong> (if the destination is unknown).
  </p>

  <h2>VLAN Database</h2>
  <div id="vlanDatabaseDisplay"></div>

  <div id="challenge">
    <h2>Packet Details</h2>
    <div id="packetDisplay"></div>
  </div>

  <form id="decisionForm">
    <label for="decision"><strong>Your Decision</strong>:</label>
    <select id="decision" required>
      <option value="">--Select Action--</option>
      <option value="forward">Forward</option>
      <option value="route">Route (Inter-VLAN)</option>
      <option value="drop">Drop</option>
      <option value="broadcast">Broadcast</option>
    </select>
    <div id="portInput" style="display:none;">
      <label for="port"><strong>Specify Port</strong> (if applicable):</label>
      <input type="number" id="port" min="1">
    </div>
    <button type="submit">Submit Decision</button>
  </form>

  <button id="newPacket" style="margin-top:20px;">New Packet Challenge</button>

  <div id="feedback" style="margin-top:20px;"></div>

  <h2>Key Terms</h2>
  <table>
    <thead>
      <tr>
        <th><strong>Term</strong></th>
        <th><strong>Definition</strong></th>
      </tr>
    </thead>
    <tbody>
      <tr>
        <td><strong>Packet</strong></td>
        <td>A formatted unit of data transmitted over a network.</td>
      </tr>
      <tr>
        <td><strong>MAC Address</strong></td>
        <td>A unique hardware identifier assigned to a network interface.</td>
      </tr>
      <tr>
        <td><strong>IP Address</strong></td>
        <td>A numerical label assigned to each device in an IP network.</td>
      </tr>
      <tr>
        <td><strong>VLAN</strong></td>
        <td>A Virtual Local Area Network that segments a network into isolated logical segments.</td>
      </tr>
      <tr>
        <td><strong>Route</strong></td>
        <td>Inter-VLAN routing; forwarding a packet between different VLANs using Layer 3 functionality.</td>
      </tr>
      <tr>
        <td><strong>Port</strong></td>
        <td>A physical or logical endpoint used to send or receive packets.</td>
      </tr>
    </tbody>
  </table>

  <script>
    /*
      **VLAN Database**: Maps VLAN IDs to objects that associate MAC addresses with ports.
      For example, in VLAN 10, MAC "AA:BB:CC:DD:EE:FF" is on Port 1.
    */
    const vlanDatabase = {
      10: {
        "AA:BB:CC:DD:EE:FF": 1,
        "22:33:44:55:66:77": 2
      },
      20: {
        "11:22:33:44:55:66": 3,
        "FF:EE:DD:CC:BB:AA": 4
      }
    };

    /*
      **Routing Table**: For inter-VLAN routing, maps an ingress VLAN and a destination VLAN
      to a specific routing port.
    */
    const routingTable = {
      10: { 20: 5 },
      20: { 10: 6 }
    };

    // A set of sample packets to challenge the student.
    const samplePackets = [
      // Ingress VLAN 10, destination MAC exists in VLAN 10: forward.
      { mac: "AA:BB:CC:DD:EE:FF", ip: "192.168.1.2", vlan: 10 },
      // Ingress VLAN 10, destination MAC exists in VLAN 20: route.
      { mac: "11:22:33:44:55:66", ip: "192.168.1.3", vlan: 10 },
      // Ingress VLAN 20, destination MAC exists in VLAN 20: forward.
      { mac: "FF:EE:DD:CC:BB:AA", ip: "192.168.1.4", vlan: 20 },
      // Ingress VLAN 20, destination MAC exists in VLAN 10: route.
      { mac: "22:33:44:55:66:77", ip: "192.168.1.5", vlan: 20 },
      // Ingress VLAN 10, unknown MAC: broadcast.
      { mac: "00:11:22:33:44:55", ip: "192.168.1.6", vlan: 10 },
      // Ingress VLAN 30, which is not in the VLAN database: drop.
      { mac: "AA:BB:CC:DD:EE:FF", ip: "192.168.1.7", vlan: 30 }
    ];

    let currentPacket = null;

    /**
     * **Decision Logic**
     * Determines the correct action for a given packet:
     *
     * 1. If the packets ingress **VLAN** is not in the **VLAN Database**, the packet is dropped.
     * 2. If the destination **MAC Address** exists in the ingress VLAN, the packet is forwarded to that port.
     * 3. If the destination **MAC Address** is found in a different VLAN, the packet should be routed via the
     *    port specified in the **Routing Table**.
     * 4. If the destination is unknown in all VLANs, the packet is broadcast.
     *
     * @param {Object} packet - The packet object with properties mac, ip, and vlan.
     * @returns {Object} An object with the correct action and, if applicable, the port.
     */
    function computeDecision(packet) {
      const ingressVLAN = packet.vlan;
      if (!vlanDatabase.hasOwnProperty(ingressVLAN)) {
        return { action: "drop" };
      }

      if (vlanDatabase[ingressVLAN].hasOwnProperty(packet.mac)) {
        return { action: "forward", port: vlanDatabase[ingressVLAN][packet.mac] };
      }

      let targetVLAN = null;
      for (let vlan in vlanDatabase) {
        if (parseInt(vlan, 10) !== ingressVLAN && vlanDatabase[vlan].hasOwnProperty(packet.mac)) {
          targetVLAN = parseInt(vlan, 10);
          break;
        }
      }

      if (targetVLAN !== null &&
          routingTable[ingressVLAN] &&
          routingTable[ingressVLAN][targetVLAN]) {
        return { action: "route", port: routingTable[ingressVLAN][targetVLAN] };
      }

      return { action: "broadcast" };
    }

    // Display the current packet details.
    function displayPacket(packet) {
      const packetDisplay = document.getElementById("packetDisplay");
      packetDisplay.innerHTML = `
        <p><strong>MAC Address</strong>: ${packet.mac}</p>
        <p><strong>IP Address</strong>: ${packet.ip}</p>
        <p><strong>Ingress VLAN</strong>: ${packet.vlan}</p>
      `;
    }

    // Render the VLAN Database so the student can see which MAC addresses are on which VLAN.
    function displayVlanDatabase() {
      const container = document.getElementById("vlanDatabaseDisplay");
      let html = '<table><thead><tr><th><strong>VLAN</strong></th><th><strong>MAC Address</strong></th><th><strong>Port</strong></th></tr></thead><tbody>';
      for (let vlan in vlanDatabase) {
        for (let mac in vlanDatabase[vlan]) {
          html += `<tr><td>${vlan}</td><td>${mac}</td><td>${vlanDatabase[vlan][mac]}</td></tr>`;
        }
      }
      html += '</tbody></table>';
      container.innerHTML = html;
    }

    // Select a random packet for the challenge.
    function newPacketChallenge() {
      currentPacket = samplePackets[Math.floor(Math.random() * samplePackets.length)];
      displayPacket(currentPacket);
      document.getElementById("feedback").innerHTML = "";
      document.getElementById("decision").value = "";
      document.getElementById("port").value = "";
      document.getElementById("portInput").style.display = "none";
    }

    newPacketChallenge();
    displayVlanDatabase();

    // Show the port input if the student selects 'forward' or 'route'.
    document.getElementById("decision").addEventListener("change", function() {
      const portInputDiv = document.getElementById("portInput");
      if (this.value === "forward" || this.value === "route") {
        portInputDiv.style.display = "block";
      } else {
        portInputDiv.style.display = "none";
      }
    });

    // Evaluate the student's decision against the computed decision.
    document.getElementById("decisionForm").addEventListener("submit", function(e) {
      e.preventDefault();
      const studentDecision = document.getElementById("decision").value;
      const studentPort = parseInt(document.getElementById("port").value, 10);

      const correct = computeDecision(currentPacket);
      let feedbackText = "";

      if (studentDecision === correct.action) {
        if (studentDecision === "forward" || studentDecision === "route") {
          if (studentPort === correct.port) {
            feedbackText = `<p style="color:green;">Correct: ${studentDecision.charAt(0).toUpperCase() + studentDecision.slice(1)} to Port ${correct.port}.</p>`;
          } else {
            feedbackText = `<p style="color:red;">Incorrect: You chose to ${studentDecision}, but the correct port is ${correct.port}.</p>`;
          }
        } else {
          feedbackText = `<p style="color:green;">Correct: ${studentDecision.charAt(0).toUpperCase() + studentDecision.slice(1)} the packet.</p>`;
        }
      } else {
        if (correct.action === "forward" || correct.action === "route") {
          feedbackText = `<p style="color:red;">Incorrect: The correct action is to ${correct.action} to Port ${correct.port}.</p>`;
        } else {
          feedbackText = `<p style="color:red;">Incorrect: The correct action is to ${correct.action} the packet.</p>`;
        }
      }
      document.getElementById("feedback").innerHTML = feedbackText;
    });

    document.getElementById("newPacket").addEventListener("click", function() {
      newPacketChallenge();
    });
  </script>
</body>
</html>
````

## Glossary

| Term | Definition |
|------|------------|
| Administrative Distance (AD) | A numerical value assigned to routing protocols to determine their trustworthiness. Lower values are considered more reliable, with directly connected routes having an AD of 0 and static routes typically having an AD of 1. |
| Area Border Router (ABR) | A router that connects one or more OSPF areas to the backbone area. It maintains separate link-state databases for each connected area and manages route summarization between areas. |
| Autonomous System (AS) | A collection of networks under a single administrative domain, typically managed by one organization. Each AS is identified by a unique number and can run its own internal routing protocols. |
| Border Gateway Protocol (BGP) | The primary routing protocol used on the internet, designed to exchange routing information between different autonomous systems. Uses path attributes and policy-based rules to determine optimal routes. |
| Bridge ID | A unique identifier used in spanning tree protocols consisting of a priority value and MAC address. Used to elect root bridges and determine network topology. |
| Broadcast Storm | A network condition where broadcast packets continuously circulate through loops in the network topology, consuming bandwidth and processing resources until the network becomes unusable. |
| Default Gateway | The router interface that devices use to send traffic destined for other networks when they don't have a more specific route. Typically configured on hosts as their exit point to other networks. |
| Designated Router (DR) | A router elected on a multi-access network segment to minimize the number of adjacencies between routers and reduce network traffic. Manages LSA flooding for its segment. |
| Enhanced Interior Gateway Routing Protocol (EIGRP) | A Cisco-developed routing protocol that combines aspects of distance vector and link-state protocols. Uses composite metrics including bandwidth and delay to determine optimal paths. |
| First Hop Redundancy Protocol (FHRP) | A category of protocols that provide redundancy for the default gateway in a network. Allows multiple routers to share a virtual IP address for fault tolerance. |
| Gateway Load Balancing Protocol (GLBP) | A Cisco protocol that provides both gateway redundancy and load balancing capabilities. Allows multiple routers to simultaneously forward traffic while acting as backups for each other. |
| Half-Duplex | A communication mode where devices can transmit and receive data, but not simultaneously. Like a walkie-talkie system where only one party can talk at a time. |
| Full-Duplex | A communication mode where devices can simultaneously transmit and receive data. Like a telephone conversation where both parties can talk at once. |
| Hot Standby Router Protocol (HSRP) | A Cisco protocol that creates a virtual router shared between two or more physical routers to provide default gateway redundancy. One router actively forwards traffic while others stand by. |
| Interior Gateway Protocol (IGP) | A class of routing protocols designed to route traffic within a single autonomous system. Includes protocols like OSPF, EIGRP, and RIP. |
| Jumbo Frame | An Ethernet frame that carries more than the standard 1500 bytes of payload, typically up to 9000 bytes. Used to improve network efficiency for large data transfers. |
| Link Aggregation | The combining of multiple physical network links into a single logical link for increased bandwidth and redundancy. Also known as port channeling or ethernet bonding. |
| Link Aggregation Control Protocol (LACP) | An IEEE standard protocol that manages the automatic bundling of physical ports into a logical link. Negotiates capabilities between devices and manages link member status. |
| Link-State Database (LSDB) | A structured collection of all link-state advertisements received by a router, representing its view of the network topology. Used by protocols like OSPF to calculate optimal routes. |
| MAC Address Table | A dynamic database maintained by switches that maps MAC addresses to physical ports. Used to make intelligent forwarding decisions for network traffic. |
| Maximum Transmission Unit (MTU) | The largest size of a single data unit that can be transmitted over a network link, typically 1500 bytes for standard Ethernet frames. Larger packets must be fragmented before transmission. |
| Multiple Spanning Tree Protocol (MST) | An extension of STP that allows different spanning-tree instances for different groups of VLANs, improving network efficiency and enabling better load balancing. |
| Native VLAN | The VLAN assigned to untagged frames received on a trunk port. Traffic from this VLAN travels untagged across trunk links while all other VLAN traffic is tagged. |
| Network Address Translation (NAT) | A process that modifies network address information in packet headers to map one address space into another, typically used to connect private networks to the internet using a single public IP address. |
| Open Shortest Path First (OSPF) | A link-state routing protocol that uses Dijkstra's shortest path first algorithm to determine optimal routes. Organizes networks into areas and supports route summarization. |
| Port Address Translation (PAT) | A type of NAT that maps multiple private IP addresses to a single public IP address using different port numbers. Also known as NAT overload or IP masquerading. |
| Port Channel | A logical interface created by aggregating multiple physical interfaces. Provides increased bandwidth and redundancy while appearing as a single interface to spanning tree and routing protocols. |
| Rapid Spanning Tree Protocol (RSTP) | An evolution of STP that provides faster convergence times through immediate state transitions and backup port roles. Typically converges in seconds rather than minutes. |
| Root Bridge | The central reference point in a spanning tree topology. All other switches calculate their shortest path to this device, and its selection influences the entire network topology. |
| Route Summarization | The process of combining multiple specific routes into a single, more general route advertisement. Reduces routing table size and improves network stability. |
| Router ID | A 32-bit number that uniquely identifies a router in routing protocols like OSPF and BGP. Often derived from IP addresses configured on the router but can be manually set. |
| Spanning Tree Protocol (STP) | A layer 2 protocol that prevents loops in switched networks by selectively blocking redundant paths while maintaining backup connectivity. |
| Static Route | A manually configured routing table entry that specifies the next hop for reaching a particular network. Requires administrative maintenance but offers precise control over routing paths. |
| Switched Virtual Interface (SVI) | A virtual interface that provides Layer 3 functionality for a VLAN in a switched network. Enables inter-VLAN routing and serves as the default gateway for VLAN members. |
| VLAN Database | A collection of VLAN configuration information stored on a switch, including VLAN IDs, names, and port assignments. Can be modified through VLAN configuration mode. |
| VLAN Tagging | The process of adding VLAN identification information to Ethernet frames, allowing multiple VLANs to share a physical link while maintaining traffic separation. |
| Voice VLAN | A specialized VLAN configured to carry voice traffic from IP phones. Enables automatic configuration of IP phones and allows a single switch port to support both a phone and a computer. |
| Virtual Router Redundancy Protocol (VRRP) | An open standard protocol that provides automatic assignment of available routers to participating hosts. Creates a virtual router that represents a group of routers sharing an IP address. |
| 802.1Q | The IEEE standard for VLAN tagging on trunk links. Defines the format for adding VLAN information to Ethernet frames and specifies how switches should handle tagged traffic. |
| Link-State Advertisement (LSA) | A packet containing routing and network information that is shared between routers in link-state protocols. Forms the basis of the link-state database used for route calculation. |
